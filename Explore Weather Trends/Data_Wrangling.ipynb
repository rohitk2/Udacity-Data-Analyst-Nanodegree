```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
df_sd = pd.read_csv('CSV/SanDiego.csv')
df_sf = pd.read_csv('CSV/SanFrancisco.csv')
df_g = pd.read_csv('CSV/global.csv')
```

```python
df = df_g.merge(
    df_sd.merge(df_sf, left_on='year', right_on='year'),
    left_on='year', right_on='year'
)
df.columns = ['Year', 'Global', 'San_Diego', 'San_Francisco']
```

```python
columns = df.columns[1:]
df_moving = df.copy()
for col in columns :
    df_moving[col] = np.array([df.iloc[(i-19):(i+1)].mean()[col] for i in range(df.shape[0])])
```

```python
plt.plot('Year', 'San_Diego', data=df);
plt.plot('Year', 'San_Francisco', data=df);
plt.plot('Year', 'Global', data=df);

plt.title('Average Temperature');
plt.xlabel('Year');
plt.ylabel('Temperature (Celcius)');
plt.legend();
```

```python
plt.plot('Year', 'San_Diego', data=df_moving);
plt.plot('Year', 'San_Francisco', data=df_moving);
plt.plot('Year', 'Global', data=df_moving);

plt.title('20 Year Moving Average Temperature');
plt.xlabel('Year');
plt.ylabel('Temperature (Celcius)');
plt.legend();
```
